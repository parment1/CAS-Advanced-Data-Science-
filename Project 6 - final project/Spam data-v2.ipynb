```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 60)
```

```python
#importing columns names
colums=[]
with open(r"C:\Users\thoma\Documents\Assignments\spam\spambase.names","r") as doc:
    lines=doc.readlines()
    for line in lines :
        colums.append(line.split(":")[0])
print(colums)
```

```python
df=pd.read_csv(r"C:\Users\thoma\Documents\Assignments\spam\spambase.data",names=colums)
df
```

```python
df.describe()
```

```python
df.isna().any().any()
```

```python
corr=df.corr()
plt.figure(figsize=(16, 12))
ax=sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns,
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True,
            linewidths=.1)
ax.set(title="all variables correlation heatmap");
```

```python
#top correlation with class (negativ and positiv)
k=10 #number of variables
cols=corr.abs().nlargest(k,'class')['class'].index
cm=np.corrcoef(df[cols].values.T)
plt.figure(figsize=(10, 8))
ax=sns.heatmap(cm,yticklabels=cols.values,
                xticklabels=cols.values,
                cmap=sns.diverging_palette(220, 10, as_cmap=True),
                annot=True,
                square=True)
ax.set(title=f"top {k} variables correlation with class heatmap");
plt.show()
```

```python
#pair plot top 4 most correlated var
sns.pairplot(data=df[cols[:5].values],hue="class",plot_kws={'alpha':0.5});
```

```python
pivot_class_freq=pd.pivot_table(df,values=df.drop(["class","capital_run_length_average","capital_run_length_longest","capital_run_length_total"],axis=1),index='class',aggfunc="mean")
pivot_class_freq
```

```python
plt.figure(figsize=(16,8))
plt.xticks(rotation=90)
sns.barplot(x=pivot_class_freq.columns,
            y=pivot_class_freq.iloc[0]-pivot_class_freq.iloc[1]).set(title="Non Spam minus Spam average word and char frequency",ylabel="average frequency difference");
```

```python
pivot_class_cap=pd.pivot_table(df,values=df[["capital_run_length_average","capital_run_length_longest","capital_run_length_total"]],index='class',aggfunc="mean")
pivot_class_cap
```

```python
plt.xticks(rotation=30)
sns.barplot(x=pivot_class_cap.columns,
            y=pivot_class_cap.iloc[0]-pivot_class_cap.iloc[1]).set(title="Non Spam minus Spam",ylabel="average difference");
```
